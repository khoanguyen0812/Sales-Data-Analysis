#### Task 4: What city had the highest number of sales?

- Let's inspect our current data really quick to see what we can use to solve this.

```python
# Inspect the table first to see what we can use
all_data.head()
```

- Oh, noticed how we do not have a city column with important values to it to figure this out? Let's make a visual representation that we can inspect.

#### Add in 'City' and 'State' column.

```python
# You can use something like this:
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

# And use .apply()

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

# To combine the 2 to look like this:

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x))
all_data['State'] = all_data['Purchase Address'].apply(lambda x: get_state(x))

all_data.head()
```

#### The city with the highest number of sales is....

```python
city_result = all_data.groupby('City').agg({'Quantity Ordered' : 'sum', 'Price Each': 'sum', 'Sales': 'sum'})
city_result_sorted = city_result.sort_values(by='Sales', ascending=False)
city_result = city_result_sorted

city_result
```

- As we can see from the sorted result, San Francisco is the city with the highest amount of sales, sitting at $8,262,203.91.

Let's put this into a graph.

```python
import matplotlib.pyplot as plt

# Calculate sales per city
city_sales = all_data.groupby('City')['Sales'].sum().sort_values(ascending=False)

# Extract cities and sales values for plotting
cities = city_sales.index.tolist()  # List of cities
sales = city_sales.values  # Corresponding sales values

# Plotting the bar graph

plt.figure(figsize=(10, 6))
plt.bar(cities, sales, color='skyblue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.title('Total Sales per City')
plt.tight_layout()

plt.show()
```

#### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data.head()
```

- From what we can draw from the table, we need to figure out a way to aggregate all of the order dates into distributions over a 24-hour period.
- Let's go ahead and convert 'order date' into date/time object.

```python
# Converting the data type and format of 'Order Date'
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], format='%m/%d/%y %H:%M')


all_data.head()
```

- Okay, so we know it works now. Let's move onto the next step. Let's add in an 'Hour' and 'Minute' column to see what we can do.

```python
# Creating new columns in the table for further calculation
all_data['Hour'] = all_data['Order Date'].dt.hour

all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

- Perfect. Let's go ahead and answer the question now.

```python
# Creating a linechart to visualize recent calculation
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').agg({'Sales':'count'}))
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

- Based on the information given by the chart, right before 11:00 a.m to 12:-- pm, and 18:00 to 19:00 seems to be a good time to display advertisement to maximize the likelihood that consumers will purchase products.

#### Question 4: What products are most often sold together?

- First, to tackle this question, let's figure out which products are being sold together.

```python
all_data.head()
```

- From what we can tell here, if the items share similar 'Order ID', it is more than likely that the items are being ordered together on the same transaction.

- Another way to think about this is that the items that are sold together will have duplicate 'Order ID', so with that said, we can utilize .count and other functions to figure out which pair of items are sold the most together.

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head(20)
```

- Okay, now we have a new dataframe with filtered information about items that are sold together indicated by the duplicated 'Order ID'.

```python
# Importing needed libraries
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2))) # Splitting the 'Grouped' items out into their own separate portion.
    
for key, value in count.most_common(10):
    print(key, value)
```

#### Question 5: What product do I think sold the most? Why do I think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')

quantity_ordered = product_group.agg({'Quantity Ordered' : 'sum'})

products = quantity_ordered.index

quantity_ordered.head()
```

- We can see here some of the quantity ordered grouped by items. Let's make this more transparent with a bar chart.

```python
plt.figure(figsize=(10, 6))

plt.bar(products, quantity_ordered['Quantity Ordered'])

plt.xticks(rotation='vertical')

plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.title('Quantity of Products Ordered')

# Display the plot
plt.tight_layout()
plt.show()
```

- From the provided graph above, AA Batterties and AAA Batteries are the most sold items out of every available products from the company.

- As to why these items are sold the most, they are more essential than the rest of the items that are offered and they are cheap.

- Let's go ahead and try to prove this by overlaying these items to see if we are right.

```python
prices = all_data.groupby('Product').agg({'Price Each': 'mean'})

print(prices)
```

```python
# Copy and paste from the other line as the template to build the secondary y-axis.
plt.figure(figsize=(10, 6))

plt.bar(products, quantity_ordered['Quantity Ordered'])

plt.xticks(rotation='vertical')

plt.xlabel('Product')
plt.ylabel('Quantity Ordered')

# Creating a secondary y-axis
plt.twinx()

# Plotting 'prices' data as a line on the secondary y-axis
plt.plot(prices.index, prices['Price Each'], color='r', marker='o', label='Average Price')

# Labels and title for the plot
plt.xlabel('Product')
plt.ylabel('Average Price', color='r')  # Secondary y-axis label
plt.title('Quantity of Products Ordered with Average Price Overlay')

# Display legends
plt.legend(loc='upper left')

# Rotate x-axis labels and adjust spacing
plt.xticks(rotation='vertical', fontsize=8)
plt.subplots_adjust(bottom=0.25)  # Adjust the bottom margin to create space for labels

# Show the combined plot
plt.tight_layout()
plt.show()
```

- With the new provided chart, red representing price and blue representing order quantity, our previous answer as to why the batteries were ordered a lot more were answered. As the listed items deemed to be more of a necessity and is significantly cheaper, the quantity ordered for that item will also significantly higher than other more expensive offered items.
